```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.utils import to_categorical

def load_mnist(images_file, labels_file):
    with open(labels_file, 'rb') as lbpath:
        labels = np.frombuffer(lbpath.read(), dtype=np.uint8, offset=8)

    with open(images_file, 'rb') as imgpath:
        images = np.frombuffer(imgpath.read(), dtype=np.uint8, offset=16).reshape(len(labels), 28, 28)

    return images, labels

X_train, y_train = load_mnist('train-images-idx3-ubyte', 'train-labels-idx1-ubyte')
X_test, y_test = load_mnist('t10k-images-idx3-ubyte', 't10k-labels-idx1-ubyte')
# Normalization
X_train = X_train.astype('float32') / 255
X_test = X_test.astype('float32') / 255

# one-hot encoding the labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# create FNN model
fnn_model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# compile, train, and evaluate FNN model
fnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
print("Training FNN model...")
fnn_history = fnn_model.fit(X_train, y_train, epochs=10, batch_size=128, validation_split=0.1)
fnn_test_loss, fnn_test_acc = fnn_model.evaluate(X_test, y_test)
print(f"FNN Test accuracy: {fnn_test_acc:.4f}")

# reshaping input for CNN
X_train_cnn = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_test_cnn = X_test.reshape(X_test.shape[0], 28, 28, 1)

# creating CNN model
cnn_model = Sequential([
    Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# compile, train, and evaluate CNN model
cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
print("\nTraining CNN model...")
cnn_history = cnn_model.fit(X_train_cnn, y_train, epochs=10, batch_size=128, validation_split=0.1)
cnn_test_loss, cnn_test_acc = cnn_model.evaluate(X_test_cnn, y_test)
print(f"CNN Test accuracy: {cnn_test_acc:.4f}")
```
